**Solution:**

I narrowed my solution down to 7 by splitting the the dataset into 4 groups of 4, then choosing the top 1 or 2 airliners from each group.

From there, I graphed the sets again, choosing my top 5 selections out of the 7 possible prospects.

Then, I made different plots until I finally came up with my answer. Hawaiian Airlines' tail falls off the before the others, but the data also reaches higher points on the graph after 0 minutes than all the rest. Alaskan Airlines has the majority of its points shifted further left than the other curves, but that may just mean they are not 'just-in-time' efficent as say US Airways. Especially since their tail is so long. 

With all that said, I choose Hawaiian Airlines. In reality, the fact that Alaskan Airlines's earlier landing times are appealing, and they are delayed less often, Hawaiian Airlines gives me the very small chance of arriving earlier when unfortunatly delayed.



```

Original code found here and modified: https://github.com/WillKoehrsen/Data-Analysis/blob/master/univariate_dist/Histogram%20and%20Density%20Plot.ipynb

# Pandas and numpy for data manipulation
import pandas as pd
import numpy as np

# Matplotlib and seaborn for plotting
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

matplotlib.rcParams['font.size'] = 18
matplotlib.rcParams['figure.dpi'] = 200

import seaborn as sns

from IPython.core.pylabtools import figsize

```

```
# Read in data
flights = pd.read_csv('formatted_flights.csv')
flights['arr_delay'].describe()

```

```
figsize(8, 4)

# List of 4 airlines to plot
airlines = ['United Air Lines Inc.', 'JetBlue Airways', 'ExpressJet Airlines Inc.',
         'Delta Air Lines Inc.']

# Iterate through airlines
for airline in airlines:
    # Subset to the airline
    subset = flights[flights['name'] == airline]
    
    # Draw the density plot
    sns.distplot(subset['arr_delay'], hist = False, kde = True,
                 kde_kws = {'linewidth': .5},
                 label = airline)
    
plt.legend(prop={'size': 6}, title = 'Airline')
plt.title('Density Plot with Multiple Airlines'); 
plt.xlabel('Delay (min)')
plt.ylabel('Density');
```

```
# Make a separate list for each airline
x1 = list(flights[flights['name'] == 'Hawaiian Airlines Inc.']['arr_delay'])
x2 = list(flights[flights['name'] == 'Alaska Airlines Inc.']['arr_delay'])
x3 = list(flights[flights['name'] == 'US Airways Inc.']['arr_delay'])
x4 = list(flights[flights['name'] == 'Delta Air Lines Inc.']['arr_delay'])
x5 = list(flights[flights['name'] == 'American Airlines Inc.']['arr_delay'])

# Assign colors for each airline and the names
colors = ['#E69F00', '#56B4E9', '#F0E442', '#009E73', '#D55E00']
names = ['Hawaiian Airlines Inc.', 'Alaska Airlines Inc.', 'US Airways Inc.',
         'Delta Air Lines Inc.', 'American Airlines Inc.']
         
# Make the histogram using a list of lists
# Normalize the flights and assign colors and names
plt.hist([x1, x2, x3, x4, x5], bins = int(180/15), normed=True,
         color = colors, label=names)

# Plot formatting
plt.legend()
plt.xlabel('Delay (min)')
plt.ylabel('Normalized Flights')
plt.title('Side-by-Side Histogram with Multiple Airlines')
```
