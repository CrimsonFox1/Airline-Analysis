```

'''Code for Plots found here and modified: https://github.com/WillKoehrsen/Data-Analysis/blob/master/univariate_dist/Histogram%20and%20Density%20Plot.ipynb'''

# Pandas and numpy for data manipulation
import pandas as pd
import numpy as np

# Matplotlib and seaborn for plotting
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline

matplotlib.rcParams['font.size'] = 18
matplotlib.rcParams['figure.dpi'] = 200

import seaborn as sns

from IPython.core.pylabtools import figsize

```

```
# Read in data
flights = pd.read_csv('formatted_flights.csv')
flights['arr_delay'].describe()

```

```
# Code block found here and Modified
sampleMeans = []
for _ in range(30):
    samples = flights['arr_delay'].sample(n=3)
    sampleMean = np.mean(samples)
    sampleMeans.append(sampleMean)
```

```
figsize(9, 7)

# Make the Probability Distribution of Arrival Delays
sns.distplot(sampleMeans, hist=True, kde=True, 
             bins=int(180/5), color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})

# Add labels
plt.title('Probability Distribution of Arrival Delays')
plt.xlabel('Delay(min)'); plt.ylabel('Probability');
```
